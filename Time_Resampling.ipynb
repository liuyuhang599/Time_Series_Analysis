```python
import pandas as pd
import numpy as np 
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('Data/starbucks.csv', index_col = 'Date', parse_dates=True)
```

```python
# daily ---> yearly
df.resample(rule='YE').mean()
```

```python
# daily ---> monthly
df.resample(rule='ME').mean()
```

```python
def first_day(entry): # Custom resampling function
    if len(entry) != 0:
        return entry[0]

df.resample(rule='ME').apply(first_day)
```
