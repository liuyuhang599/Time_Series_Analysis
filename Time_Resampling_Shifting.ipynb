```python
import pandas as pd
import numpy as np 
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('Data/starbucks.csv', index_col = 'Date', parse_dates=True)
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.index
```

```python
# daily ---> yearly
df.resample(rule='YE').mean()
```

```python
# daily ---> monthly
df.resample(rule='ME').mean()
```

```python
def first_day(entry): # Custom resampling function
    if len(entry) != 0:
        return entry[0]

df.resample(rule='ME').apply(first_day)
```

```python
df['Close'].resample(rule='QE').mean().plot.bar() # quarterly
```

```python
df.shift(1)
```

```python
df.shift(5)
```

```python
df.shift(periods=2, freq='D')
```

## Rolling and Expanding

```python
df['Close'].plot(figsize=(12,5))
```

```python
df['Close'].plot(figsize=(12,5))
df.rolling(window=30, min_periods=5).mean()['Close'].plot()
```

```python
df['Close'].plot(figsize=(12,5))
df['Close'].expanding().mean().plot(figsize=(12,5))
```
